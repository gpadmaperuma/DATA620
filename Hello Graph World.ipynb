```python
names = {0:"Andre", 1:"Beverly", 2:"Carol", 3:"Diane", 4:"Ed", 5:"Fernando", 6:"Garth", 7:"Heather", 8:"Ike", 9:"Jane"}
names
```

```python
# Creating the graph object for names
G1=nx.Graph()

# Create edges manually
e1 = [(0, 1), (0, 2), (0, 3), (0, 5), (1, 3), (1, 4), (1, 6), (2, 3), (2, 5), 
      (3, 4), (3, 5), (3, 6), (4, 6), (5, 6), (5, 7), (6, 7), (7, 8), (8, 9)]

# Add all the edges to the graph object G1
G1.add_edges_from(e1)

print("Edges from the new network", "\n", G1.edges())
```

```python
# Add all the nodes (names) to the graph object G1
G1 = nx.relabel_nodes(G1, names)

print("Nodes from the new network", "\n", G1.nodes())
```

```python
# Print the graph
nx.draw(G1, with_labels=names)
plt.show()
```
