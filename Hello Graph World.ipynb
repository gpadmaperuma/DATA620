First step is to import necessary libraries to generate these graphs. I will be using matplotlib package to generate the plots and networkx to study the network. **%matplotlib inline** is our magic command for IPython that allows you to add plots to the browser interface.

```python
import matplotlib.pyplot as plt 
import networkx as nx 
%matplotlib inline
```

Below command will create an empty with no nodes and edges. In Networkx, nodes can be any hashable object such as a text string, an image, a XML object etc. 

```python
# creating an empty grapgh.
G=nx.Graph()
```

```python
# generate krackhardt kite graph
g=nx.generators.small.krackhardt_kite_graph()
```

```python
# printing nodes and edges of the krackhardt_kite_graph
print("Nodes from krackhardt_kite_graph", "\n", g.nodes(), "\n")
print("Edges from krackhardt_kite_graph","\n", g.edges())
```

```python
print("Number of Edges in this network", "\n", g.number_of_edges())
```

```python
print("Number of Nodes in this network", "\n", g.number_of_nodes())
```

```python
nx.draw(g, with_labels=1)
plt.show()
```

```python
# create nodes and labeling nodes as people
names = {0:"Andre", 1:"Beverly", 2:"Carol", 3:"Diane", 4:"Ed", 5:"Fernando", 6:"Garth", 7:"Heather", 8:"Ike", 9:"Jane"}
names
```

```python
# Creating the graph object for names
G1=nx.Graph()

# Create edges manually - using the same connection as in krackhardt kite graph
e1 = [(0, 1), (0, 2), (0, 3), (0, 5), (1, 3), (1, 4), (1, 6), (2, 3), (2, 5), 
      (3, 4), (3, 5), (3, 6), (4, 6), (5, 6), (5, 7), (6, 7), (7, 8), (8, 9)]

# Add all the edges to the graph object G1
G1.add_edges_from(e1)

print("Edges from the new network", "\n", G1.edges())
```

```python
# Add all the nodes (names) to the graph object G1
G1 = nx.relabel_nodes(G1, names)

print("Nodes from the new network", "\n", G1.nodes())
```

```python
# Print the graph
nx.draw(G1, with_labels=names)
plt.show()
```
